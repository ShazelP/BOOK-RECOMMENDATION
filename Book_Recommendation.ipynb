**MOTIVATION**


> In the era of digital content proliferation, the sheer volume of books available to users has grown exponentially. Amidst this abundance of information, recommending the right book to the right user has become a critical challenge. Traditional book recommendation systems have primarily relied on collaborative filtering, which uses user behavior data to make recommendations. However, this approach faces limitations, particularly when dealing with new users or new books with little to no prior data. To address these challenges, this project aims to develop a content-based recommendation model that analyzes book descriptions, calculates the similarity between books, and recommends related books based on this analysis.

\
**OBJECTIVE**


> The primary objective of this project is to build a recommendation model that suggests books similar to a user’s current selection, based on the content of book descriptions. By leveraging the information contained in these descriptions, the model will automatically identify and recommend books that share thematic or content-related similarities with the user’s chosen book. The goal is to enhance the user’s reading experience by offering more personalized and relevant book suggestions, enabling them to discover new books that align with their interests.


**ASSUMPTIONS**


> **1. Importance of Book Descriptions:**\
It is assumed that book descriptions effectively encapsulate the key themes and content of a book, making them a valuable resource for determining the similarity between books. The similarity of these descriptions is presumed to be a reliable indicator of books that users might be interested in.\
**2. Appropriateness of TF-IDF Vectorization:**\
The project assumes that TF-IDF vectorization is a suitable method for converting book descriptions into vectors that capture the significance of keywords within each book. The cosine similarity calculated from these vectors is expected to be an effective measure for quantifying the content-based similarity between books.\
**3. Effectiveness of Similarity-Based Recommendations**:\
It is assumed that the similarity between books, as determined by their descriptions, will serve as a meaningful criterion for making book recommendations. By utilizing cosine similarity to assess the relatedness of book descriptions, the model is expected to provide recommendations that are closely aligned with the content of the user’s current selection.

# 3. TEXT PREPROCESSING

```python
#Preprocess the descriptions by removing all characters except Korean ones
def Remove_but_KR(text):
    pattern = r'[^ㄱ-ㅎㅏ-ㅣ가-힣]\s'
    result = re.sub(pattern, '', text)

    result = re.sub(r'[&%$#@!《》()[]{}"\dA-Za-z]', '', result)

    return result


#apply
df["BOOK_INTRCN_CN"] = df["BOOK_INTRCN_CN"].apply(Remove_but_KR)
df
```

```python
#define stopwords
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','을','를','으로','자','것','에','와','한','하다']
#tokenize into morphemes
okt = Okt()
```

```python
#remove stopwords
def Remove_stopwords(text):
    if isinstance(text, list):  # if it's list
        text = ' '.join(text)
    tokens = okt.morphs(text, stem=True)
    result = [word for word in tokens if not word in stopwords]
    return ' '.join(result)

#apply
df["BOOK_INTRCN_CN"] = df["BOOK_INTRCN_CN"].apply(Remove_stopwords)
df
```

```python
#check
df["BOOK_INTRCN_CN"][:10]
```

# 4. VECTORIZATION AND CALCULATION OF SIMILARITY

```python
#vectorize after tf-idf
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df["BOOK_INTRCN_CN"])
tfidf_matrix.shape
```

```python
#change data type to reduce memory usage
tfidf_matrix = tfidf_matrix.astype(np.float32)
```

```python
#cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
#Reset the index based on book titles.
indices = pd.Series(df.index, index=df["TITLE_NM"])
```

# 5. BUILD BOOK RECOMMENDATION MODEL

```python
#make function for book recommendation
def get_rocomended_books(title, cosine_sim=cosine_sim):
    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx])) #list cosine similarity by books
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True) #descend order
    sim_scores = sim_scores[1:6] #extract top 5
    book_indices = [i[0] for i in sim_scores] #extracted index
    return df.iloc[book_indices][["TITLE_NM", "AUTHR"]]
```
